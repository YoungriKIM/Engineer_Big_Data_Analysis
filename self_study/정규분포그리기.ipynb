```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
# 정규분포 값 지정
mean = 0
std = 1
```

```python
x = np.linspace(-4*std + mean, 4*std + mean, 1000)
pdf = ss.norm.pdf(x, mean, std)
```

```python
plt.figure(figsize=(8, 5)) # 그림 크기 설정
plt.plot(x, pdf, label='Mean = {}, Std = {}'.format(mean, std)) # 그래프 그리기
plt.title('Normal Distribution') # 제목
plt.xlabel('X') # x축 라벨
plt.ylabel('Probability Density') # y축 라벨
plt.legend() # 범례 표시
plt.grid(True) # 격자 보이기
plt.show() # 그래프 보여주기
```
