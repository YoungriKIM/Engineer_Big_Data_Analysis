```python
import numpy as np                  # numpy 패키지를 np라는 이름으로 불러옴

ar1 = np.array([1,2,3,4,5])         # 리스트 [1,2,3,4,5]를 ndarray 객체 ar1으로 생성
ar2 = np.array([[1,2,3],[4,5,6]])   # 2차원 배열 생성
print(ar1)      # ar1의 내용 출력
print(ar2)      # ar2의 내용 출력
```

```python
import numpy as np                  # numpy 패키지를 np라는 이름으로 불러옴

ar3 = np.random.randn(3)            # 3 크기의 1차원 배열을 생성(난수 발생)
ar4 = np.random.randn(2,3)          # 2*3 크기의 2차원 배열을 생성(난수 발생)
print(ar3)
print(ar4)
```

```python
import numpy as np

ar5 = np.zeros(5)       # 1차원 배열 [0,0,0,0,0] 생성
ar6 = np.ones((2,3))  # 2차원 배열 [[1,1,1][1,1,1]] 생성
print(ar5)
print(ar6)
```

```python
import numpy as np

ar7 = np.arange(20, 200, 10)     # 20부터 200-1까지 10씩 증가
ar8 = ar7.reshape(3,6)           # 3*6 배열로 재배치
print(ar7)             # 배열 ar7 출력
print(ar8)             # 배열 ar8 출력
```

```python
import numpy as np

ar1 = np.arange(1, 21, 1)               # 1부터 20(21-1)까지 1씩 증가
print(ar1)                              # 배열 ar1 출력
print(ar1[1])                           # 배열의 두번째 원소 출력(인덱스는 0부터 시작)
ar2 = ar1.reshape(2,10)                 # 2*10 배열로 재구성
print(ar2)                              # 배열 ar1 출력
ar2[1][1] = 100                         # ar2[1][1]의 원소를 100으로 변경
print(ar2[1][0], ar2[1][1], ar2[1][2])  # ar2[1][0], ar2[1][1], ar2[1][2]의 원소 출력
```

```python
ar1 = np.arange(1, 11, 1)         # 1부터 10(11-1)까지 1씩 증가
ar2 = ar1 + 3                     # ar1 각각의 원소에 3을 더함
ar3 = ar1 * 2                     # ar1 각각의 원소에 2를 곱함
print(ar1)                        # 배열 ar1 출력
print(ar2, "\n", ar3)           # 배열 ar2, \n(줄바꿈 기호), ar3 출력
```

```python
ar1 = np.array([[5,7,9],[-7,-6,19],[6,9,11]])
print(ar1)
print(ar1.sum())           # 배열 원소의 합계
print(ar1.mean())          # 배열 원소의 평균
print(ar1.max())           # 배열 원소 중 가장 큰 값
print(ar1.min())           # 배열 원소 중 가장 작은 값
print(ar1.max(axis=0))     # 배열의 각 열의 원소 중 가장 큰 값을 구함
print(ar1.max(axis=1))     # 배열의 각 행의 원소 중 가장 큰 값을 구함
```

```python
ar1 = np.array([[5,7,9],[-7,-6,19],[6,9,11]])
print(ar1 > 0)                        # ar1의 각 원소가 0보다 큰지 판단한 결과를 출력
m_count = (ar1 < 0).sum()             # 원소 중 음수가 몇 개인지 계산 (True는 1, False는 0)
print(m_count)

# ar1의 원소가 0보다 작으면 0으로 변경, 그렇지 않으면 ar1의 값으로 대입
ar2 = np.where(ar1 < 0, 0, ar1)
print(ar2)
```

```python
ar1 = np.array([[5,7,9], [-3,-6,19], [6,4,11]])
print(ar1)      # ar1 출력
ar1.sort(0)     # sort(0)는 열 단위로 원소를 정렬
print(ar1)
ar1.sort(1)     # sort(1)는 행 단위로 원소를 정렬
print(ar1)
```

>>**02 판다스(Pandas) 데이터프레임**

```python
import pandas as pd                     # 판다스 패키지를 불러와서 pd라는 이름으로 사용

sr1 = pd.Series([10, 30, 20, 40, 60])   # 리스트를 시리즈 객체로 생성
print(sr1)
```

```python
sr1 = pd.Series([10, 30, 20, 40, 60])   # 리스트를 시리즈 객체로 생성
print(sr1.values)                       # 시리즈 sr1의 값 출력
print(sr1.index)                        # 시리즈 sr1의 인덱스 출력 (처음값, 마지막값, 증가값)
print(sr1[2])                           # 인덱스 2에 해당하는 값 출력
```

```python
sr1 = pd.Series([10,30,20,40,60], index=['a','b','c','d','e'])  # 문자열 인덱스 설정
print(sr1[1])                                                   # 시리즈 sr1[1]의 값 출력 - 숫자 인덱스
print(sr1['b'])                                                 # 시리즈 sr1['b']의 값 출력 - 문자열 인덱스
```

```python
sr1 = pd.Series([10,30,20,40,60], index=['a','b','c','d','e'])
print(sr1[[1,2]])                               # 시리즈 sr1[1], sr1[2]의 값 출력
print(sr1[['a','c']])                       # 시리즈 sr1['a'], sr1['c']의 값 출력
```

```python
sr1 = pd.Series([10,30,20,40,60], index=['a','b','c','d','e'])
print(sr1[1:4])               # sr1[1]부터 sr1[3]까지 출력
print(sr1['b':'d'])           # sr1['b']부터 sr1['d']까지 출력
```
