```python
# housing 데이터 사용
# median_house_value를 종속 변수로 설정
# 상관성 분석 후 필요한 변수만 사용
# 평균제곱오차(MSE) 로 평가하기
# RandomForestRegressor 사용
# 결측치 처리
    # 결측값 행 삭제
    # 변수형 데이터 삭제
# 분할 7:3
```

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# housing 데이터 사용
df = pd.read_csv('C:/Engineer_Big_Data_Analysis/DATAS/이기적데이터/housing.csv')
df
```

```python
# 결측치 처리
    # 결측값 행 삭제
    # 변수형 데이터 삭제
df.dropna(inplace=True)
df.drop(columns='ocean_proximity', axis=1, inplace=True)
```

```python
# 상관성 분석 후 필요한 변수만 사용
df.corr(method='pearson')
# 종속 변수 제외 모두 사용
```

```python
# 데이터 준비
X = df[['longitude', 'latitude', 'housing_median_age', 'total_rooms','total_bedrooms', 'population', 'households', 'median_income']]
y = df['median_house_value']

# 분할 7:3
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=311)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# 모델 준비
model = RandomForestRegressor()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# 값이 높으니 x_train으로 확인
mean_squared_error(y_train, model.predict(x_train))
```
