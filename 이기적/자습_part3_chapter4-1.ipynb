```python
# K-means 이용하기
    # 데이터셋에 K개의 centroids를 지정
# iris 데이터 이용하기
    # 범주형 데이터 숫자로 인코딩
# 라벨이 없다는 가정하에 k-means를 이용하기
# 상관성 시각화 해보기
```

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# 데이터 불러오기
df = pd.read_csv('C:/Engineer_Big_Data_Analysis/DATAS/이기적데이터/iris.csv')
df
```

```python
# 범주형 데이터는 숫자로 바꿔주기
print(df.dtypes)
print('-'*50)

encoder = LabelEncoder()
df['species'] = encoder.fit_transform(df['species'])

print(df.dtypes)
```

```python
# 데이터셋 준비
# 라벨이 없다는 가정이기 때문에 X만 준비
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# 클러스터 준비
cluster = KMeans(n_clusters = 3, # 클러스터 개수
                 n_init = 10, # 초기 entroid 개수
                 max_iter = 500, # 알고리즘 반복 횟수
                 random_state=311)
# 클러스터 학습
cluster.fit(X)
```

```python
# 각 군집의 중심점 확인하기
cluster.cluster_centers_
```

```python
# 클러스터의 예측 결과 보기
print(cluster.predict(X))

# 예측값을 원래 df에 추가하기
df['cluster'] = cluster.predict(X)
df
```
