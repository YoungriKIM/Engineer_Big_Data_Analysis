```python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

display(train.head(3))
print(train.shape)
display(test.head(3))
print(test.shape)
```

```python
# 전처리
# 1. 결측치 | 필수
# 2. 이상치 | 문제에서 명시하면
# 3. 인코딩 | 범주형 데이터가 있다면 필수
# 4. 스케일링 | 선택
```

```python
# 결측치는 삭제 혹은 대체?
# 1차로 삭제해서 모델을 만들어서 성능을 보고, 튜닝을 하면서 채우는 것을 추천
```

```python
# 결측치 삭제
print(train.shape)
df = train.dropna()
print(df.shape)
### test 데이터는 삭제하면 절대 안 됨에 주의

# 특정 컬럼 기준으로 결측치 행 삭제
df = train.dropna(subset=['native.country', 'workclass'])
print(df.shape)

# 혹은 결측치가 있는 컬럼을 삭제
df = train.dropna(axis=1)
print(df.shape)
```

```python
# 결측치 채우기

# 1) 범주형에 최빈값으로 채우기
m = train['workclass'].mode()[0]
train['workclass'] = train['workclass'].fillna(m)
m = train['native.country'].mode()[0]
train['native.country'] = train['native.country'].fillna(m)

train.isna().sum()
```

```python
# 2) 임의의 값으로 대체
train['occupation'] = train['occupation'].fillna('X')
train.isna().sum()
```

```python
# test에도 동일한 결측치 처리
print(test.isna().sum().sum())
test['workclass'] = test['workclass'].fillna(train['workclass'].mode()[0])
test['native.country'] = test['native.country'].fillna(train['native.country'].mode()[0])
test['occupation'] = test['occupation'].fillna('X')
print(test.isna().sum().sum())
```

```python
# 3) 수치형 데이터에 평균값으로 대체하기
mean_age = int(train['age'].mean())
train['age'] = train['age'].fillna(mean_age)
test['age'] = test['age'].fillna(mean_age)
print(mean_age)

# 4) 수치형 데이터에 중앙값으로 대체하기
median_hpw = int(train['hours.per.week'].median())
train['hours.per.week'] = train['hours.per.week'].fillna(median_hpw)
test['hours.per.week'] = test['hours.per.week'].fillna(median_hpw)
print(median_hpw)
```

```python
# 결측치 없는지 확인
train.isna().sum()
test.isna().sum()
```

```python
# 이상치 처리
# 이번에는 age만 다뤄봄

display(train[train['age'] <= 0])
display(test[test['age'] <= 0]) # test에는 없음

print(train.shape)
train = train[train['age'] > 0]
print(train.shape)
```

```python
# 인코딩 (범주형 데이터 숫자화)y

# 전에 y_train은 분리해서 인코딩 대상에서 빼기
### 주목 ### 컬럼 분리하면서 동시에 빼기
y_train = train.pop('income')
```

```python
# 원핫 인코딩 pd.get_dummies(train)
# 를 쓰기 전에 주의할 점
# train - test 를 합치지 않고 적용하면?
print(train.shape)
aft_train = pd.get_dummies(train)
print(aft_train.shape)

print(test.shape)
aft_test = pd.get_dummies(test)
print(aft_test.shape)
print('-'*50)

# train은 컬럼이 107로 test는 103개로 분리되어 학습에 사용할 수 없다.
# >> 이런 경우 concat하여 한 번에 get_dummies
concat_df = pd.concat([train, test])
print(concat_df.shape)
encoded_concat_dt = pd.get_dummies(concat_df)
print(encoded_concat_dt.shape)
encoded_train = encoded_concat_dt.iloc[:len(train)].copy()
encoded_test = encoded_concat_dt.iloc[len(train):].copy()
print(encoded_train.shape)
print(encoded_test.shape)
```

```python
# 레이블에 인코딩하기

# object인 컬럼만 찾고
cols = train.select_dtypes(include='object').columns
print(cols)

# 사이킷런으로 레이블인코더 적용
from sklearn.preprocessing import LabelEncoder

# train으로 fit_fransform 하고
# test에 transform 만
for colname in cols:
    le = LabelEncoder()
    train[colname] = le.fit_transform(train[colname])
    test[colname] = le.transform(test[colname])
    
display(train.sample(5))
train.dtypes
```

```python
# 수치형 데이터 스케일링 하기
# 스케일링은 1차 제출 후 시간적 여유가 있다면 적용하고 비교를 추천
cols = ['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']

# 세 가지 스케일링 기법 비교를 위해 df 복사 함수 만들고
def get_data():
    train_copy = train.copy()
    test_copy = test.copy()
    return train_copy, test_copy

# 스케일링은 인코더처럼 한 개의 컬럼씩 진행하지 않아도 된다
```

```python
# 1 민맥스 스케일링
# 데이터를 0~1 사이로 변환 (가장 작은 값이 0, 가장 큰 값이 1)
train_copy, test_copy = get_data()

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

display(train_copy[cols].head(2))
train_copy[cols] = scaler.fit_transform(train_copy[cols])
test_copy[cols] = scaler.transform(test_copy[cols])
display(train_copy[cols].head(2))
```

```python
# 2 스탠더드 스케일링
# 평균이 0, 표준편차가 1로 만드는 방법

train_copy, test_copy = get_data()

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

display(train_copy[cols].head(2))
train_copy[cols] = scaler.fit_transform(train_copy[cols])
test_copy[cols] = scaler.transform(test_copy[cols])
display(train_copy[cols].head(2))
```

```python
# 로버스트 스케일링
# 각 값의 중앙값을 빼고 1사분위와 3사분위의 차이로 나누는 방법
# 다른 스케일링에 비해 이상치에 덜 영향을 받음

from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

display(train[cols].head(2))
train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])
display(train[cols].head(2))
```

```python
# target인 y가 문자여서 인코딩하고싶다면?
# map이나 replace를 활용 (여기서는 안 함)
y_train_copy = y_train.copy()

y_train_copy.map({'>50K':0, '<=50K':1})
y_train_copy.replace('>50K', 0).replace('<=50K', 1)
```

```python
# 검증 데이터 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(train, y_train, test_size=0.2, random_state=311)
x_train.shape, x_val.shape, y_train.shape, y_val.shape
```

```python
# 머신러닝 학습 및 평가
# 기본적으로 트리 모델을 쓸 건데, 과적합을 방지하기 위해서 앙상블 모델을 사용할 거임
# Random Forest, LightGBM 을 추천

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(x_train, y_train)
pred_proba = rf.predict_proba(x_val)
pred = rf.predict(x_val)

# target에 들어간 숫자를 확인
print(rf.classes_)
print(pred_proba[:3])
print(pred[:3])
```

```python
# 평가 지표 이용하기
```

```python
# ROC-AUC
# roc_auc 사용하기 > 확률값으로 반환하는 predict_proba가 필요함
# 또한 roc_auc는 양성(1) 라벨을 기준으로 하기 때문에 proba의 결과에서 1번째를 잘라와서 사용

from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_val,pred_proba[:, 1]) # 정답 먼저
print('roc_auc: ', roc_auc)
```

```python
# ACC ascore

from sklearn.metrics import accuracy_score

acc = accuracy_score(y_val, pred)
print('acc: ', acc)
```

```python
# F1 score
# F1 score는 타겟이 인코딩되어있으면 그냥 데이터를 넣고
# 아니라면 양성 값이 무엇인지 명시해줘야 함 (.classes_ 참고)

from sklearn.metrics import f1_score

f1 = f1_score(y_val, pred, pos_label = '>50K')
print('f1: ', f1)
```

```python
# LightGBM 모델로 사용하기
import lightgbm as lgb

lgbmc = lgb.LGBMClassifier(random_state=0, verbose=-1)
lgbmc.fit(x_train, y_train)

pred_proba = lgbmc.predict_proba(x_val)
pred = lgbmc.predict(x_val)

from sklearn.metrics import roc_auc_score, accuracy_score, f1_score
# roc_auc
roc_auc = roc_auc_score(y_val, pred_proba[:, 1])
print('roc_auc:', roc_auc)
# acc
acc = accuracy_score(y_val, pred)
print('acc:', acc)
# f1 score
f1 = f1_score(y_val, pred, pos_label = '>50K')
print('f1:', f1)
```

```python
# 예측 및 결과 파일 생성

# test를 활용해서 예측하고 파일 저장하기
pred = lgbmc.predict_proba(test)
print(pred)

submit = pd.DataFrame({'pred': pred[:, 1]})
display(submit.head(3))
submit.to_csv('result.csv', index=False)
```
