```python
# 결측치는 삭제 혹은 대체?
# 1차로 삭제해서 모델을 만들어서 성능을 보고, 튜닝을 하면서 채우는 것을 추천
```

```python
# 결측치 삭제
print(train.shape)
df = train.dropna()
print(df.shape)
### test 데이터는 삭제하면 절대 안 됨에 주의

# 특정 컬럼 기준으로 결측치 행 삭제
df = train.dropna(subset=['native.country', 'workclass'])
print(df.shape)

# 혹은 결측치가 있는 컬럼을 삭제
df = train.dropna(axis=1)
print(df.shape)
```

```python
# 결측치 채우기

# 1) 범주형에 최빈값으로 채우기
m = train['workclass'].mode()[0]
train['workclass'] = train['workclass'].fillna(m)
m = train['native.country'].mode()[0]
train['native.country'] = train['native.country'].fillna(m)

train.isna().sum()
```

```python
# 2) 임의의 값으로 대체
train['occupation'] = train['occupation'].fillna('X')
train.isna().sum()
```

```python
# test에도 동일한 결측치 처리
print(test.isna().sum().sum())
test['workclass'] = test['workclass'].fillna(train['workclass'].mode()[0])
test['native.country'] = test['native.country'].fillna(train['native.country'].mode()[0])
test['occupation'] = test['occupation'].fillna('X')
print(test.isna().sum().sum())
```

```python
# 3) 수치형 데이터에 평균값으로 대체하기
mean_age = int(train['age'].mean())
train['age'] = train['age'].fillna(mean_age)
test['age'] = test['age'].fillna(mean_age)
print(mean_age)

# 4) 수치형 데이터에 중앙값으로 대체하기
median_hpw = int(train['hours.per.week'].median())
train['hours.per.week'] = train['hours.per.week'].fillna(median_hpw)
test['hours.per.week'] = test['hours.per.week'].fillna(median_hpw)
print(median_hpw)
```

```python
# 결측치 없는지 확인
train.isna().sum()
test.isna().sum()
```

```python
# 이상치 처리
# 이번에는 age만 다뤄봄

display(train[train['age'] <= 0])
display(test[test['age'] <= 0]) # test에는 없음

print(train.shape)
train = train[train['age'] > 0]
print(train.shape)
```

```python
# 인코딩 (범주형 데이터 숫자화)y

# 전에 y_train은 분리해서 인코딩 대상에서 빼기
y_train = train.pop('income')
```

```python
# 원핫 인코딩 pd.get_dummies(train)
# 를 쓰기 전에 주의할 점
print(train.shape)
aft_train = pd.get_dummies(train)
print(aft_train.shape)

print(test.shape)
aft_test = pd.get_dummies(test)
print(aft_test.shape)
print('-'*50)

# train은 컬럼이 107로 test는 103개로 분리되어 학습에 사용할 수 없다.
# >> 이런 경우 concat하여 한 번에 get_dummies
concat_df = pd.concat([train, test])
print(concat_df.shape)
encoded_concat_dt = pd.get_dummies(concat_df)
print(encoded_concat_dt.shape)
encoded_train = encoded_concat_dt.iloc[:len(train)].copy()
encoded_test = encoded_concat_dt.iloc[len(train):].copy()
print(encoded_train.shape)
print(encoded_test.shape)
```

```python
# 레이블에 인코딩하기

# object인 컬럼만 찾고
cols = train.select_dtypes(include='object').columns
print(cols)

# 사이킷런으로 레이블 인코딩하기
from sklearn.preprocessing import LabelEncoder

for colname in cols:
    le = LabelEncoder()
    train[colname] = le.fit_transform(train[colname])
    test[colname] = le.transform(test[colname])
    
display(train.sample(5))
```
