```python
# 라이브러리 호출, 데이터 불러오기
import pandas as pd

train = pd.read_csv('diabetes_train.csv')
test = pd.read_csv('diabetes_test.csv')

print(train.shape, test.shape)
```

```python
# 라이브러리 호출, 데이터 불러오기
import pandas as pd

train = pd.read_csv('diabetes_train.csv')
test = pd.read_csv('diabetes_test.csv')
target = train.pop('Outcome')
print(train.shape, test.shape)

train.head()
```

```python
# minmax scaler 사용해보기
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

train = scaler.fit_transform(train)
test = scaler.transform(test)
```

```python
# 데이터 분리
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
```

```python
# roc_auc 지표 사용
from sklearn.metrics import roc_auc_score

# 랜포클래시파이어 사용
# 하이퍼파라미터 수정해보기
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0,
                             n_estimators=500,
                             max_depth=20,
                             )

rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_val)

print(roc_auc_score(y_val, y_pred))

# 0.7178082191780822
# 0.7209589041095891
```

```python
pred = rfc.predict_proba(test)[:, -1]
result = pd.DataFrame({'pred':pred})
result.to_csv('result.csv', index=False)

# 제출 파일 확인
pd.read_csv('result.csv')
```

#### 문제2 | 이직 여부 예측

```python
# train : hr_train.csv
# test : hr_test.csv
# 예측할 컬럼 target(0 잔존, 1 이직)

# 예측값은 이직할 확률
# 제출은 result.csv ,pred컬럼명
# 채점은 roc_auc
```

```python
# 라이브러리 호출, 데이터 불러오기
import pandas as pd

train = pd.read_csv('hr_train.csv')
test = pd.read_csv('hr_test.csv')
print(train.shape, test.shape)
```

```python
# 데이터 미리보기
display(train.head(3))
display(test.head(3))
```

```python
# 데이터 타입보기
print(train.info())
print('-'*50)
print(test.info())
```

```python
# 결측치 확인
print(train.isna().sum())
print('-'*50)
print(test.isna().sum())
```

```python
# 데이터 분포 확인
display(train.describe())
display(test.describe())
```

```python
# 오브젝트 타입도 확인
display(train.describe(include='O'))
display(test.describe(include='O'))
```

```python
# 데이터 전처리
```

```python
# 데이터 합쳐서 진행
target = train.pop('target')
print(train.shape, test.shape)
df = pd.concat([train,test])
print(df.shape)
```

```python
# 결측치 처리

print(sum(df.isna().sum()))

# 1
# 모두 최빈값으로 처리
# isna_col_list = ['gender', 'enrolled_university', 'education_level', 'major_discipline', 'experience', 'company_size', 'company_type', 'last_new_job']
# for isna_col_name in isna_col_list:
#     df[isna_col_name] = df[isna_col_name].fillna(df[isna_col_name].mode()[0])

# 2
# 모두 'X'로 채우기
df = df.fillna('X')

print(sum(df.isna().sum()))
```

```python
# 인코딩

# 오브젝트 타입 컬럼은 모두 적용
o_col_list = df.select_dtypes(include='O').columns.to_list()

display(df.head(3))
from sklearn.preprocessing import LabelEncoder
for o_col_name in o_col_list:
    le = LabelEncoder()
    df[o_col_name] = le.fit_transform(df[o_col_name])
display(df.head(3))
```

```python
# 스케일링 진행
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
scaler = RobustScaler()
origin_col_list = df.columns.to_list()
df = pd.DataFrame(scaler.fit_transform(df), columns=origin_col_list)
```

```python
# 상관관계 확인하기
display(df.corrwith(target).sort_values(ascending=False))

# 상관성 낮은 컬럼 삭제
print(df.shape)
df = df.drop(columns=['major_discipline', 'gender', 'last_new_job'])
print(df.shape)
```

```python
# train, test 분리하고 
# train - val 분리
train = df.iloc[:len(train)].copy()
test = df.iloc[len(train):].copy()
print(train.shape, test.shape)

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
```

```python
# 모델 학습 및 검증
```

```python
# 랜포 모델 사용
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0,
                             n_estimators=220,
                             max_depth=10)

# roc_auc 지표 사용
from sklearn.metrics import roc_auc_score

rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_val)
print(roc_auc_score(y_val, y_pred))
```

```python
# 예측 및 제출 
```

```python
pred = rfc.predict_proba(test)[:,-1]

print(pred)
```

```python
result = pd.DataFrame({'pred':pred})
result.to_csv('result.csv', index=False)
```
