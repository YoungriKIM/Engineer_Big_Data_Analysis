```python
# 14 
import numpy as np
df = pd.read_csv('type1_data1.csv')

# 중복 데이터 제거
print(df.shape)
df.drop_duplicates(inplace=True)
print(df.shape)

# f3 컬럼의 결측치는 0, silver는 1, gold는 2, vip는 3으로 변환
def change_f3(x):
    if x == 'silver': x=1
    elif x == 'gold': x=2
    elif x == 'vip': x=3
    else: x=0
    return x
df['f3'] = df['f3'].apply(change_f3)

# 변환된 f3의 총합을 정수로
df['f3'].sum()
```

```python
# 16
df = pd.read_csv('type1_data1.csv')

# views 컬럼의 1사분위수, 3사분위수 그리고 IQR 구하기
quantile_1 = df['views'].quantile(0.25)
quantile_3 = df['views'].quantile(0.75)
iqr = quantile_3 - quantile_1
print(quantile_1, quantile_3, iqr)

# 이상치 조건에 맞는 데이터 찾기
# 이상치는
#   1사분위수 - (IQR * 1.5) 보다 작은 값
#   3사분위수 + (IQR * 1.5) 보다 큰 값
min_out = quantile_1 - (iqr * 1.5)
max_out = quantile_3 + (iqr * 1.5)
out_df = df[(df['views'] < min_out)|(df['views'] > max_out)]
display(out_df)

# 이상치 데이터의 views 컬럼 합을 정수로 반환 
int(sum(out_df['views']))
```

```python
# 17
df = pd.read_csv('type1_data1.csv')

# views 컬럼의 표준편자
df_std = df['views'].std()
print(df_std)

# age 컬럼의 이상치를 제거하고, views 컬럼의 표준편차 다시 구하기
# 소수점 나이, 음수, 0
df.dropna(axis=0, subset='age', inplace=True)
df = df[df['age'] > 0]
df = df[df['age']%1 == 0.0]
new_df_std = df['views'].std()
print(new_df_std)

# 이상치 제거 전 후의 views 컬럼의 표준편차를 더하고 반올림 소수 둘째자리까지
round(df_std+new_df_std, 2)
```

```python
# 19
df = pd.read_csv('type1_data1.csv')

# 결측치를 바로 뒤에 있는 값으로 대체하기(바로 뒤 값이 결측치일 경우, 뒤에 있는 데이터 중 가장 가까운 값으로 대체)
print(df.isna().sum().sum())
df.fillna(method='bfill', inplace=True)
print(df.isna().sum().sum())

# city와 f2 컬럼을 기준으로 그룹합을 계산하고
# views 값이 세 번째로 큰 city 이름 구하기
print(df.groupby(by=['city', 'f2']).sum().sort_values('views', ascending=False).iloc[2].name[0])
```
