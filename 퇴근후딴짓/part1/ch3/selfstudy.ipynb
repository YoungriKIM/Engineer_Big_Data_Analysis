```python
# 01

import pandas as pd
df = pd.read_csv('type1_data1.csv')
df
```

```python
# f5 컬럼이 0이 아닌 데이터(행)
target_df = df[df['f5']!=0]
display(target_df.head(3))
print('='*70)

# 위 df에서 views 결측치를 views 최솟값으로 채우기
print('views 결측치 개수', sum(target_df['views'].isna()))
target_df['views'] = target_df['views'].fillna(target_df['views'].min())
display(target_df.head(3))
print('views 결측치 개수', sum(target_df['views'].isna()))
print('='*70)

# views 컬럼 중앙값 계산해 정수로 반환
### 주목 ###
# median 중앙값
int(target_df['views'].median())
```

```python
# 3
df = pd.read_csv("type1_data1.csv")
print(df.shape)

# 결측치가 있는 행 제거
df.dropna(axis=0, inplace=True)
print(df.shape)

# views 컬럼을 f1 칼럼으로 나눈 값을 새로운 컬럼으로 추가
df['new_col'] = df['views']/df['f1']

# 새로운 컬럼의 값 중 가장 큰 값을 가진 행을 age을 정수로 반환
int(df.loc[df['new_col'].idxmax(), 'age'])
```

```python
# 4
df = pd.read_csv("type1_data1.csv")

# views 컬럼의 결측 데이터를 0으로 대체
print(sum(df['views'].isna()))
df['views'] = df['views'].fillna(0)
print(sum(df['views'].isna()))

# views 컬럼에서 상위 10번째 값 구하기
tenth = df['views'].sort_values(ascending=False).reset_index().loc[9, 'views']
print(tenth)

# views 컬럼에서 상위 10개의 값을 상위 10번째 값을 대체하기
target_index = df['views'].sort_values(ascending=False).reset_index().loc[0:9, 'index'].to_list()
print(target_index)

df.loc[target_index, 'views'] = tenth
df['views'].loc[target_index]

# views 컬럼의 전체 합을 정수로 반환
int(sum(df['views']))
```

```python
# 5
df = pd.read_csv("type1_data1.csv")

# f4 컬럼 중에서 'FJ'가 포함된 데이터 찾기
### 주목 ###
# 특정 문자열 포함하는 데이터 찾을 때는 먼저 str 다음에 contains 생각
df[df['f4'].str.contains('FJ') == True]
display(df[df['f4'].str.contains('FJ') == True])

# 찾은 데이터에서 'f2'의 평균값을 구하기(반올림 소수 둘째 자리까지)
round(df[df['f4'].str.contains('FJ') == True]['f2'].mean(), 2)
```

```python
# 6
df = pd.read_csv("type1_data1.csv")

# f3 컬럼이 gold 이면서 f2 컬럼이 2 인 데이터
display(df[(df['f3'] == 'gold')&(df['f2']==2)])

#  찾은 데이터의 f1 컬럼의 분산을 반환, 반올림 후 소수 둘째 자리까지
round(df[(df['f3'] == 'gold')&(df['f2']==2)]['f1'].var(), 2)
```

```python
# 7
df = pd.read_csv("type1_data1.csv")

# 모든 나이(age)에 1을 더하기
df['age'] = df['age']+1

# 20대의 views 평균과 30대의 views 평균의 절댓값 차이를 구하기(반올림 후 소수 둘째 자리)
v_mean_20 = df[(df['age'] >= 20)&(df['age'] < 30)]['views'].mean()
v_mean_30 = df[(df['age'] >= 30)&(df['age'] < 40)]['views'].mean()
round(abs(v_mean_20 - v_mean_30), 2)
```

```python
# 8
df = pd.read_csv("type1_data1.csv")

# 방법1
# subscribed 컬럼이 2024년 2월인 데이터 찾기
df['subscribed'] = pd.to_datetime(df['subscribed'])
dt_df = df[(df['subscribed'].dt.month == 2)&(df['subscribed'].dt.year == 2024)]
display(dt_df)

# 방법2
df = pd.read_csv("type1_data1.csv")
str_df = df[df['subscribed'].str[:7] == '2024-02']
display(str_df)

# 찾은 데이터 중 f3 컬럼이 gold인 데이터의 개수
print(len(dt_df[dt_df['f3'] == 'gold']))
print(len(str_df[str_df['f3'] == 'gold']))
```
