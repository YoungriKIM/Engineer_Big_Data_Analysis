```python
# 16
df = pd.read_csv('type1_data1.csv')

# views 컬럼의 1사분위수, 3사분위수 그리고 IQR 구하기
quantile_1 = df['views'].quantile(0.25)
quantile_3 = df['views'].quantile(0.75)
iqr = quantile_3 - quantile_1
print(quantile_1, quantile_3, iqr)

# 이상치 조건에 맞는 데이터 찾기
# 이상치는
#   1사분위수 - (IQR * 1.5) 보다 작은 값
#   3사분위수 + (IQR * 1.5) 보다 큰 값
min_out = quantile_1 - (iqr * 1.5)
max_out = quantile_3 + (iqr * 1.5)
out_df = df[(df['views'] < min_out)|(df['views'] > max_out)]
display(out_df)

# 이상치 데이터의 views 컬럼 합을 정수로 반환 
int(sum(out_df['views']))
```

```python
# 25
df = pd.read_csv('delivery_time.csv')

# 각 사용자별로 첫 주문과 마지막 주문 사이의 시간 간격을 일 단위로 계산하기
df['주문시간'] = pd.to_datetime(df['주문시간'])
df = df[['user','주문시간']]
target_df = df.groupby('user')['주문시간'].agg(['min', 'max'])
target_df['diff'] = (target_df['max'] - target_df['min']).dt.days
display(target_df.head(3))

# 시간 간격이 1일 이하인 사용자를 제외하고 나머지 사용자들의 평균 시간 간격(일 단위) 계산
mean_diff = target_df[target_df['diff'] > 1]['diff'].mean()
print(mean_diff)

# 평균 시간 간격보다 긴 시간 간격을 가진 사용자의 수를 정수로 반환
len(target_df[target_df['diff']>mean_diff])
```

```python
# 29
df = pd.read_csv('delivery_time.csv')

# 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산
    # 5km 미만: 2000
    # 5km 이상 10km 미만: 4000
    # 10km 이상 15km 미만: 6000
    # 15km 이상 20km 미만: 8000
df = df[['주문시간', '거리']]
def price_of_d(x):
    if x < 5: x=2000
    elif 5 <= x < 10: x=4000
    elif 10 <= x < 15: x=6000
    elif 15 <= x < 20: x=8000
    return x
df['배달료'] = df['거리'].apply(price_of_d)
    
# 월별로 배달료의 총합을 집계
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['주문월'] = df['주문시간'].dt.month
df.groupby('주문월')['배달료'].sum()

# 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 반환
print(df.groupby('주문월')['배달료'].sum().idxmax())
print(int(df.groupby('주문월')['배달료'].sum().max()))
```

```python
# 33
df = pd.read_csv('school_data.csv')

# 과목에 상관없이 점수가 가장 작은 점수 25개 찾기
### 주목 ### 여러 열의 정보를 한 줄로 길게 늘이기: .melt(id_vars='기준컬럼', value_vars='내릴컬럼')
df = df.melt(id_vars='이름', value_vars=['수학', '영어', '국어'])
### 주목 ### n 개의 작거나 큰 값 보기 : nsmallest(n), nlargest(n)
df['value'].nsmallest(25)

# 찾음 점수 25개의 합을 정수로
int(sum(df['value'].nsmallest(25)))
```
