```python
# 개인 신용 관련 정보 데이터를 사용해서 개인의 신용 등급(Credit_Score)를 예측
# 평가 기준은 macro-f1
# 타겟은 신용 등급으로 1,2,3 세 종류
# 제출 방식은 test 데이터로 예측한 1개 컬럼만 csv로 제출(컬럼명 pred, 파일명 result.csv)
```

```python
# 라이브러리 호출, 데이터 불러오기
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# EDA
```

```python
# 간단하게 확인
print(train.shape, test.shape)
display(train.head(3))
display(test.head(3))
```

```python
# 결측치 확인
print(train.isna().sum())
print(test.isna().sum())
# 없는 상태
```

```python
# 데이터 타입 확인
print(train.dtypes)
print('-'*50)
print(test.dtypes)
# object는 3개
```

```python
# 데이터 분포 확인
display(train.describe())
display(test.describe())
```

```python
# object도 확인
display(train.describe(include='O'))
display(test.describe(include='O'))
```

```python
# target 데이터도 확인
train['Credit_Score'].value_counts()
```

```python
# 데이터 전처리
```

```python
# target 분리
print(train.shape, test.shape)
target = train.pop('Credit_Score')
print(train.shape, test.shape)
```

```python
# 인코딩
from sklearn.preprocessing import LabelEncoder

# train, test 합쳐서 한 번에 진행
df = pd.concat([train,test])
print(df.shape)

# 인코딩할 columns 이름 리스트
col_list = df.select_dtypes('O').columns.to_list()

for col_name in col_list:
    le = LabelEncoder()
    df[col_name] = le.fit_transform(df[col_name])
    
print(df.shape)
df.head(3)
```

```python
# train, test 나누기
print(len(train), len(test))
train = df.iloc[:len(train)].copy()
test = df.iloc[len(train):].copy()
print(len(train), len(test))
```

```python
# 검증 데이터 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=2024)
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
```

```python
# 모델 학습 및 평가
```

```python
# 평가에 acc, f1 사용하기
from sklearn.metrics import accuracy_score, f1_score

# LightGBM 사용하기
import lightgbm as lgb
lgbm = lgb.LGBMClassifier(random_state=2024, verbose=-1)

lgbm.fit(x_train, y_train)
y_pred = lgbm.predict(x_val)

print(accuracy_score(y_val, y_pred))
print(f1_score(y_val, y_pred, average='macro'))
```

```python
# 예측 및 결과 생성
```

```python
# 예측하고
pred = lgbm.predict(test)

# df / 컬럼명 'pred' / 파일명 result.csv / index=False
pred_df = pd.DataFrame(pred, columns=['pred'])
pred_df.to_csv('result.csv', index=False)
```
