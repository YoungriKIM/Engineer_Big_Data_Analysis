```python
train = pd.read_csv('score_train.csv')
test = pd.read_csv('score_test.csv')
```

```python
target = train.pop('Credit_Score')
print(train.shape, test.shape)
```

```python
# 결측치 없음
```

```python
# 인코딩
from sklearn.preprocessing import LabelEncoder
df = pd.concat([train, test])

for col_name in df.select_dtypes(include='O').columns.to_list():
    le = LabelEncoder()
    df[col_name] = le.fit_transform(df[col_name])
```

```python
# 데이터 분리
train = df.iloc[:len(train)].copy()
test = df.iloc[len(train):].copy()

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
```

```python
# 모델 학습 및 검증
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

model = RandomForestClassifier(random_state=666,
                               n_estimators=80,
                               max_depth=10,
                              #  class_weight='balanced'
                               )

model.fit(x_train, y_train)
y_pred = model.predict(x_val)
print(f1_score(y_val, y_pred, average='macro'))

# 0.6853171856067161
# 0.6853171856067161
# 0.6976879614450985
# 0.7050710604920153
```

```python
# 예측 및 제출
```

```python
pred = model.predict(test)
result = pd.DataFrame({'pred':pred})
result.to_csv('result.csv', index=False)

pd.read_csv('result.csv')
```

#### 문제2 | 약물 종류 예측

```python
# 데이터 : drug_train.csv, drug_test.csv
# 예측값: Drug(DrugY, drugX, drugA, drugC, drugB)
# 평가지표: f1-macro
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import f1_score

train = pd.read_csv('drug_train.csv')
test = pd.read_csv('drug_test.csv')

print(train.shape, test.shape)

# EDA
# train.head(3), test.head(3)
# test.info()
# train.isna().sum(), test.isna().sum()
# display(train.describe(include='O'))
# display(test.describe(include='O'))

# 전처리
target = train.pop('Drug')
print(train.shape, test.shape)
df = pd.concat([train, test])
print(df.shape)

# 결측치 없음
# 인코딩
for col_name in df.select_dtypes(include='O').columns.to_list():
    le = LabelEncoder()
    df[col_name] = le.fit_transform(df[col_name])

# 분리
train = df.iloc[:len(train)].copy()
test = df.iloc[len(train):].copy()
print(train.shape, test.shape)
x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)

# 모델 학습 및 검증
model = RandomForestClassifier(random_state=0
                               , n_estimators=8
                               , max_depth=12
                               )
model.fit(x_train, y_train)
y_pred = model.predict(x_val)
print(f1_score(y_val, y_pred, average='macro'))

# 예측 및 저장
pred = model.predict(test)
print(pred)
# result = pd.DataFrame({'pred':pred})
# result.to_csv('result.csv')
```

#### 문제3 | 유리 종류 예측

```python
# 데이터 : glass_train.csv, glass_test.csv
# 예측값 : Type(1,2,3,5,6,7)
# 평가지표 : f1-weighted
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import f1_score

train = pd.read_csv('glass_train.csv')
test = pd.read_csv('glass_test.csv')
print(train.shape, test.shape)
display(train.head(2))
display(test.head(2))

# EDA
# train.info()
# test.info()
# train.describe()
# 결측치 없음, 데이터 분포 양호함
train['Type'].value_counts()
# 타겟값 불균형

# 데이터 분리
target = train.pop('Type')
print(train.shape, test.shape)
x_train, x_val, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

# 모델 학습 및 검증
# model = RandomForestClassifier(random_state=0, class_weight='balanced')
model = LGBMClassifier(random_state=0, verbose=-1, class_weight='balanced', n_estimator=100, max_depth=12)
model.fit(x_train, y_train)
y_pred = model.predict(x_val)
print(f1_score(y_val, y_pred, average='weighted'))

# 예측 및 제출
pred = model.predict(test)
result = pd.DataFrame({'pred':pred})
result.to_csv('result.csv', index=False)
pd.read_csv('result.csv')
```
