## 작업형1

### 문제1

```python
import pandas as pd
# df = pd.read_csv("data4-1.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-1.csv")

# 1) 3사분위수와 1사분위수의 차이를 절대값으로 계산
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
result = abs(q1 - q3)

# 2) 소수점 이하를 버리고, 정수로 출력
print(int(result))
```

### 문제2

```python
import pandas as pd
# df = pd.read_csv("data4-2.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv")

# 1) 비율이 40%보다 크고 50%보다 작은 조건
cond1 = (df['loves'] + df['wows']) / df['reactions'] > 0.4
cond2 = (df['loves'] + df['wows']) / df['reactions'] < 0.5

# 2) type이 video인 조건
cond3 = df['type'] == 'video'

# 3) 조건에 맞는 데이터 수 구하기
print(len(df[cond1 & cond2 & cond3]))
```

### 문제3

```python
# 방법1
import pandas as pd
# df = pd.read_csv("data4-3.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-3.csv")

# 1) datetime으로 형변환
df['date_added'] = pd.to_datetime(df['date_added'])

# 2) dt를 활용해 year과 month 파생변수 생성
df['year'] = df['date_added'].dt.year
df['month'] = df['date_added'].dt.month

# 3) 조건
cond1 = df['country'] == "United Kingdom"
cond2 = df['year'] == 2018
cond3 = df['month'] == 1

# 4) 조건에 맞는 데이터 수 출력
print(len(df[cond1 & cond2 & cond3]))
```

```python
# 방법2
# df = pd.read_csv("data4-3.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-3.csv")

# 1) datetime으로 형변환
df['date_added'] = pd.to_datetime(df['date_added'])

# 2) 조건
cond1 = df['country'] == "United Kingdom"
cond2 = df['date_added'] >= '2018-1-1'
cond3 = df['date_added'] <= '2018-1-31'

# 3) 조건에 맞는 데이터 수 출력
print(len(df[cond1 & cond2 & cond3]))
```
