### GroupBy
GroupBy can be summarized as Split-Apply-Combine.
* [데이터 출처](https://github.com/justmarkham)

```python
import pandas as pd
from IPython.display import display

drinks_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(drinks_url)
print(drinks.shape)
drinks.head(3)

# beer 맥주, spirit 증류수, wine 와인, total litres 순수 알코올 양, continent 속한 대륙
```

```python
# 평균 맥주 소비량보다 많은 양을 소비하는 continent
# 맥주 소비량 평균 구해놓기
beer_avrg = drinks.groupby("continent").beer_servings.mean().mean()
print(f"continent기준 맥주 소비량 평균: {beer_avrg:.3f}")
# continent 기준 그룹핑
group_continent_df = pd.DataFrame(drinks.groupby("continent").beer_servings.mean())
# beer_avrg 보다 높은 continent 확인
group_continent_df[group_continent_df['beer_servings']>beer_avrg]
```

```python
# 각 대륙별 와인 소비에 대한 통계량 확인
drinks.groupby('continent').wine_servings.describe()
```

```python
# 각 열에 대한 대륙별 평균 알코올 소비량 확인
display(drinks.groupby('continent').mean(numeric_only=True))
print('-'*50)

# 각 열에 대한 대륙별 알코올 소비량 중간값 확인
drinks.groupby('continent').median(numeric_only=True)
```

```python
# 대륙별 spirit consumption 통계값에 대한 mean, min, max 값 확인
drinks.groupby('continent').spirit_servings.agg(['mean','min','max'])
```
