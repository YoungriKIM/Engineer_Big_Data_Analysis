```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(url)
print(tips.shape)
tips.head(3)
```

```python
# Unnamed 0 열 삭제
del tips['Unnamed: 0']
print(tips.shape)
tips.head(3)
```

```python
# total_bill 열을 histogram으로 그리기
ttbill = sns.histplot(tips.total_bill)

# 상세 설정
ttbill.set(xlabel='Value',
           ylabel='Frequency',
           title='Total Bill')

# 상단과 우측의 선을 없애서 깔끔하게 보기
sns.despine()
```

```python
# total_bill과 tip 사이의 관계를 볼 수 있는 산점도 그리기
sns.jointplot(x="total_bill", y='tip', data=tips)
```

```python
# total_bill, tip, size에 대한 각각의 관계를 확인하기 위한 그래프 그리기
# 셋다 숫자형임
sns.pairplot(tips)
```

```python
# day와 total_bill 관계를 볼 수 있는 스트립플롯 그리기
# jitter > 데이터 포인트를 수평으로 약간씩 흩뜨려서 표시할지
sns.stripplot(x='day',
              y='total_bill',
              data=tips,
              jitter=True)
```

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True)
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = tips)
```

```python
# better seaborn style
sns.set(style = "ticks")

# creates FacetGrid
g = sns.FacetGrid(tips, col = "time")
g.map(plt.hist, "tip");
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7)

g.add_legend();
```
