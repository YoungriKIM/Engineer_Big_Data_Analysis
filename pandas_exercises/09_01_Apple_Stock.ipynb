```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 데이터 불러오기
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
apple = pd.read_csv(url)
print(apple.shape)
apple.head(3)
```

```python
# 각 열의 데이터 타입 확인
apple.dtypes
```

```python
# Date 열을 datetime type 으로 만들고 인덱스로 지정하기
apple.Date = pd.to_datetime(apple.Date)
apple.set_index('Date', inplace=True)
apple.head()
```

```python
# 날짜에 중복값이 있는지 확인
print(f'중복 개수 값: {apple.index.duplicated().sum()}')
apple.index.is_unique
```

```python
# 데이터는 최신순으로 정렬되어 있음, 오름차로 재정렬
apple.sort_index(ascending=True, inplace=True)
apple
```

```python
# 매월 마지막 영업일의 평균 가격보기
# BME = Business Month End, 매월 마지막 평일
apple.resample('BME').mean()
```

```python
# 데이터 중 가장 오래된 날짜와 가장 최근 날짜 사이의 차이는 몇일인지
print(apple.index.min())
print(apple.index.max())
print('-'*50)
(apple.index.max() - apple.index.min()).days
```

```python
# 우리가 가진 데이터에는 총 몇개의 월이 있는지
apple.resample('ME').mean().shape[0]
```

```python
# Adj Close에 대한 플롯차트 그리기
# 사이즈는 20 * 5 인치로
apple_adjc = apple['Adj Close'].plot(title='Apple Stock')
# 상세 설정
fig = apple_adjc.get_figure()
fig.set_size_inches(20, 5)
```
