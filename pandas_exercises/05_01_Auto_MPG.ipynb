### MPG Cars
* [데이터 출처](https://archive.ics.uci.edu/dataset/9/auto+mpg)

```python
# 라이브러리 호출
import pandas as pd
import numpy as np
from IPython.display import display

# 데이터 가져오기
cars1_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
cars2_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv"

cars1 = pd.read_csv(cars1_url)
cars2 = pd.read_csv(cars2_url)

print(cars1.shape)
display(cars1.head(3))
print(cars2.shape)
display(cars2.head(3))
```

```python
# cars1 에 unnamed 데이터가 있으니 그것을 제외하기
cars1 = cars1.loc[:, 'mpg':'car']
cars1.head(3)
```

```python
# 각 데이터셋의 개수 확인
print(f"cars1 데이터 개수: {cars1.shape[0]}\ncars2 데이터 개수: {cars2.shape[0]}")
```

```python
# 두 개의 데이터를 'cars' 이름으로 합치기
cars = pd.concat([cars1, cars2])
print(cars.shape)
cars.head(3)
```

```python
# 'owner' 이름의 15000~73000 사이 랜덤수가 들어간 열 만들기
# dtype l은 long으로 int64와 동일
    # 'i': 32비트 정수 (np.int32)
    # 'l': 64비트 정수 (np.int64)
    # 'f': 32비트 부동 소수점 (np.float32)
    # 'd': 64비트 부동 소수점 (np.float64)
cars['owner'] = np.random.randint(low='15000', high='73001', size=cars.shape[0], dtype='l')
cars.head(3)
```
