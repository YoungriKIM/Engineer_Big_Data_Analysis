### Iris
* [데이터 출처](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data)

```python
# 라이브러리 호출
import pandas as pd
import numpy as np

# 데이터 불러오기
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)
iris
```

```python
# 컬럼이름 지정해주기
col_names = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris.columns = col_names
iris
```

```python
# 데이터에 빈값이 있는지 확인
iris.isnull().sum()
# 없음!
```

```python
# 10:29행의 petal_length 값음 NaN으로 지정하기
iris.loc[10:29, 'petal_length'] = np.nan
# 확인
iris.isnull().sum()
```

```python
# NaN값을 1.0 으로 바꾸기
# fillna 사용
# 열을 지정해서 바꾸거나
iris.petal_length.fillna(1.0)
# df 전체를 바꾸기
iris.fillna(1.0, inplace=True)
# 확인
iris.isnull().sum()
```

```python
# class 열 삭제
del iris['class']
iris
```

```python
# 0:3 행의 값을 NaN으로 바꾸기
iris.iloc[0:3] = np.NaN
iris
```

```python
# NaN을 가진 행  삭제하기
iris.dropna(inplace=True)
iris
```

```python
# Index를 0부터 시작하도록 다시 지정하기
iris.reset_index(drop=True, inplace=True)
iris
```
