```python
# 라이브러리 호출
import pandas as pd
import numpy as np
from IPython.display import display

# 주소로 데이터 불러오기
data_address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(data_address, sep = '|', index_col='user_id')
```

```python
# 처음 4개 확인
display(users.head(4))

# 끝 7개 확인
display(users.tail(7))
print('-'*50)

# 데이터 행,열 크기 확인
print(f"행: {users.shape[0]}, 열: {users.shape[1]}")
print('-'*50)

# 데이터 열 이름 확인하기
print(users.columns)
print('-'*50)

# 데이터 인덱스 확인
print(users.index)
```

```python
# 각 열의 데이터 타입 확인
print(users.dtypes)
```

```python
# occupation 열 데이터 출력하기
print('--- 방법1 ---')
print(users.occupation)
print('--- 방법2 ---')
print(users['occupation'])
```

```python
# 직업 종류는 총 몇개인지
print(f"직업 종류: {users['occupation'].nunique()} 가지")
print('-'*50)

# 가장 많았던 직업 종류 5가지 표로 보기
print(users['occupation'].value_counts().head())
print('-'*50)

# 첫 번째 직업명만 확인하기
users['occupation'].value_counts().index[0]
```

```python
# 데이터 정보 요약하기

# 기본은 숫자형 데이터만 요약
print("describe default")
display(users.describe())
print('-'*50)

# 모든 열 정보 요약하기
print("include all")
display(users.describe(include="all"))
print('-'*50)

# 오프젝트(스트링) 형식만 요약하기
print("include object")
display(users.describe(include="object"))
print('-'*50)

# zip_code만 요약하기
print("zip_code describe")
users.zip_code.describe()
```

```python
# 나이 평균값(반올림)
print(round(users.age.mean()))
print('-'*50)

# 가장 적은 비율로 수집된 나이 확인
# age 기준으로 그룹핑
print(users.age.value_counts())
print('-'*50)
# 가장 적은 횟수 확인
min_count = users.age.value_counts().unique()[-1]
print(min_count)
print('-'*50)
# 가장 적은 횟수가 값인 나이만 확인
users.age.value_counts()[users.age.value_counts()==min_count]
```
