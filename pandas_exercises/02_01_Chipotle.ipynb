Filtering and Sorting Data
* chipo 데이터 다시 사용

```python
# 라이브러리 호출
import pandas as pd
from IPython.display import display

# 데이터 불러오기
chipo_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(chipo_url, sep = '\t')
```

```python
# $10.00 초과의 상품은 몇 개인지

# item_price float 형으로 변환
f_item_price = [float(i[1:]) for i in chipo['item_price']]
chipo['item_price'] = f_item_price

# item_name, quantity, choice_description 에 중복값 삭제
print(f"중복값 삭제 전 행 개수: {chipo.shape[0]}")
chipo_filtered = chipo.drop_duplicates(['item_name', 'quantity', 'choice_description'])
print(f"중복값 삭제 후 행 개수: {chipo_filtered.shape[0]}")

# quantity가 1인 데이터만 남기기
chipo_one_prod = chipo_filtered[chipo_filtered['quantity'] == 1]
print(f"quantity 1인 행 개수: {chipo_one_prod.shape[0]}")
```

```python
# 방법1
# 표로 확인
display(chipo_one_prod[chipo_one_prod['item_price']>10])
# item_name만 반환(중복 제외)
print(list(chipo_one_prod[chipo_one_prod['item_price']>10]['item_name'].unique()))
print(len(chipo_one_prod[chipo_one_prod['item_price']>10]['item_name'].unique()))
print('-'*50)

# 방법2
# 표로 확인
display(chipo_one_prod.query('item_price > 10'))
# item_name만 반환(중복 제외)
print(list(chipo_one_prod.query('item_price > 10')['item_name'].unique()))
print(len(chipo_one_prod.query('item_price > 10')['item_name'].unique()))
```

```python
# 각 아이템의 가격을 확인하기 행은 item_name, item_price 행만 남길 것

# quantity가 1인 것만 남기기
chipo_q_1 = chipo[chipo['quantity']==1]
# item_name, choice_description 기준으로 중복 제거
chipo_unique = chipo_q_1.drop_duplicates(['item_name','choice_description'])
# 아이템 이름만 미리 뽑아놓기
chipo_pro_name_list = list(chipo_unique['item_name'].unique())
print(len(chipo_pro_name_list), chipo_pro_name_list)
print('-'*50)
# chipo_unique item_name, choice_description만 남기고, 인덱스 정리하기
chipo_unique = chipo_unique[['item_name', 'item_price']].reset_index(drop=True)

# 각 상품별로 돌면서 가장 저렴한 가격의 index 저장
pro_index_list = []
for pro_name in chipo_pro_name_list:
    # display(chipo_unique[chipo_unique['item_name']==pro_name].sort_values(by='item_price'))
    target_index = chipo_unique[chipo_unique['item_name']==pro_name].sort_values(by='item_price').iloc[0].name
    pro_index_list.append(target_index)
# 저장한 index만 남기고 가격 내림차 정렬
chipo_filtered = chipo_unique.iloc[pro_index_list].sort_values(by='item_price').reset_index(drop=True)

# 상품 개수, 가격정보, 표 확인
print(f"총 상품 종류: {chipo_filtered.shape[0]}\n최저가 상품: {chipo_filtered.iloc[0][0]} {chipo_filtered.iloc[0][1]}\n최고가 상품: {chipo_filtered.iloc[-1][0]} {chipo_filtered.iloc[-1][1]}")
display(chipo_filtered)
```

```python
# 데이터를 item_name 기준으로 정렬하기
chipo.sort_values(by='item_name', ascending=True)
```

```python
# 금액이 가장 비싼 주문의 주문 수량 보기
display(chipo.sort_values(by='item_price', ascending=False).head(5))
chipo.sort_values(by='item_price', ascending=False).iloc[0]['quantity']
```
