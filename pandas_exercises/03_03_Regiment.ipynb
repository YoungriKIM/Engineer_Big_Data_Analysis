```python
# 라이브러리 호출
import pandas as pd
from IPython.display import display

# 데이터 입력
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

```python
# df로 저장
regiment = pd.DataFrame(raw_data)
print(regiment.shape)
regiment.head(3)
```

```python
# regiment가 Nighthawks인 데이터의 preTestScore mean값
regiment[regiment['regiment'] == 'Nighthawks'].preTestScore.mean()
```

```python
# company별 기본적인 통계값 확인
regiment.groupby('company').describe()
```

```python
# 각 company별 preTestScore 평균값
regiment.groupby('company').preTestScore.mean()
```

```python
# regiment와 company 기준의 그룹의 평균 preTestScore
# type 확인
print(type(regiment.groupby(['regiment', 'company']).preTestScore.mean()))
# 데이터 확인
regiment.groupby(['regiment', 'company']).preTestScore.mean()
```

```python
# regiment와 company 기준의 그룹의 평균 preTestScore, 대신 인덱스값을 컬럼으로 올려서 df type 으로 확인
# type 확인
print(type(regiment.groupby(['regiment', 'company']).preTestScore.mean().unstack()))
# 데이터 확인
regiment.groupby(['regiment', 'company']).preTestScore.mean().unstack()

# stack은 열을 행으로, unstack은 행을 열으로 올리는 기능
```

```python
# 모든 데이터를 regiment와 company 기준으로 그룹핑
regiment.groupby(['regiment', 'company']).describe()
```

```python
# 각각의 regiment, company는 몇개의 데이터를 갖고 있는지
# 방법1 - count는 NaN 값을 제외
print(regiment.groupby(['regiment', 'company']).count()['name'])
print('-'*50)
# 방법2 - size는 NaN 값도 포함
regiment.groupby(['regiment', 'company']).size()
```

```python
# regiment로 그룹핑한 데이터를 각 regiment 별로 name, data를 출력
# groupby한 데이터 구조 먼저 확인
display(pd.DataFrame(regiment.groupby('regiment')))
# 0열은 regiment name이 1열은 해당 regiment의 모든 데이터로 된 구조

# for문 돌면서 이름, 데이터 출력하기
for name, wholedata in regiment.groupby('regiment'):
    print('▶', name)
    print(wholedata)
    print('-'*50)
```
