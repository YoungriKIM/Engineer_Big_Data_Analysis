### GroupBy
* [데이터 출처](https://github.com/justmarkham)

```python
import pandas as pd
from IPython.display import display

users_url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(users_url, sep='|', index_col='user_id')
print(users.shape)
users.head(3)
```

```python
# 직업별 평균 연령 확인하기
users.groupby('occupation').age.mean()
```

```python
# 직업별 남성 비율 구하고, 높은 순으로 정렬
# 성별 숫자로 변환할 함수 만들고
def gender_to_num(x):
    if x == 'M' : return 1
    else : return 0
# 적용해서 gender_num 열 만들기
users['gender_num'] = users.gender.apply(gender_to_num)

# gender_num 합에서(M이 1이니 sum은 1의 개수) 각 직업별 총 표본수를 나누기 (Series/Series)
male_per_occu = users.groupby('occupation').gender_num.sum() / users.groupby('occupation').gender_num.count() * 100

# 내림차 정렬
male_per_occu.sort_values(ascending=False)
```

```python
# 각 직업별로 min, max 나이 확인
users.groupby('occupation').age.agg(['min','max'])
```

```python
# 각 직업,성별 별로 평균 나이 구하기
users.groupby(['occupation', 'gender']).age.mean()
```

```python
# 각 직업별 여성, 남성 비율
# 성별별 개수 count / 전체 개수 count 로 비율 계산
users.groupby(['occupation', 'gender']).gender.count() / users.groupby('occupation').gender.count() * 100
```
