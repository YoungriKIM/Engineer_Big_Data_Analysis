### Pokemon
* 데이터 직접 만들어서 사용하기, 포켓몬이라니 너무 귀엽잖아

```python
import pandas as pd

raw_data = {"name": ['이상해씨', '파이리', '꼬부기', '캐터피'],
            "evolution": ['이상해풀', '리자드', '어니부기', '단데기'],
            "type": ['grass', 'fire', 'water', 'bug'],
            "hp": [45, 39, 44, 45],
            "pokedex": ['yes', 'no','yes','no']                      
            }

pokemon = pd.DataFrame(raw_data)
pokemon
```

```python
# column 순서가 알파벳순서로 들어갔으니 원하는 대로 다시 지정하기
pokemon = pokemon[['name', 'type', 'hp', 'evolution', 'pokedex']]
pokemon
```

```python
# place 열 추가하기
pokemon = pokemon.copy()
pokemon['place'] = ['park','street','lake','forest']
pokemon
```

```python
# 각 열의 데이터 타입보기
pokemon.dtypes
```

```python
# attack, defence, speed 열 추가하기
pokemon['attack'] = [49, 52, 48, 30]
pokemon['defence'] = [49, 43, 65, 35]
pokemon['speed'] = [45, 65, 43, 45]
# 총합 열 추가
pokemon['stats_total'] = pokemon['hp'] + pokemon['attack'] + pokemon['speed']
pokemon
```

```python
# 열 순서 바꾸기
pokemon = pokemon.reindex(columns=['name', 'evolution', 'type', 'place', 'pokedex', 'hp','attack', 'defence', 'speed', 'stats_total'])
pokemon
```
