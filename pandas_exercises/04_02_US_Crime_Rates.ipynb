```python
# 라이브러리 호출
import pandas as pd
import numpy as np
from IPython.display import display

# 데이터 가져오기
csv_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(csv_url)
print(crime.shape)
crime.head(3)
```

```python
# 컬럼들의 타입 확인
print(crime.dtypes)
print('-'*50)
# year 칼럼의 타입은 int64임을 확인

# 타입을 datetime64로 바꾸기
crime.Year = pd.to_datetime(crime.Year, format='%Y')
display(crime.head(3))

# 컬럼들의 타입 확인
print(crime.dtypes)
```

```python
# Year을 index로 지정하기
crime.set_index('Year',inplace=True)
crime.head(3)
```

```python
# Total 컬럼 삭제하기
del crime['Total']
crime.head(3)
```

```python
# 10년 단위로 그룹으로 묶고 합계 확인하기 - datetime 형식이 index일 때 적용할 수 있는 resample 기능을 사용

# 리샘플링 주기 규칙
    # 'YE': 연간
    # 'Q': 분기별
    # 'M': 월간
    # 'W': 주간
    # 'D': 일간
    # 'H': 시간별
    # 'T' 또는 'min': 분별
    # 'S': 초별

# 리샘플링 후 데이터 집계
    # .mean(): 평균
    # .sum(): 합계
    # .max(): 최대값
    # .min(): 최소값
    # .median(): 중앙값
    # .ohlc(): 시가, 고가, 저가, 종가
    # .first(): 첫 번째 값
    # .last(): 마지막 값

# 다른 값은 sum을 사용하지만, Polulation은 가장 큰 값으로 지정해야 함
crimes_10 = crime.resample('10YE').sum()
display(crimes_10.head())

# population만 max로 따로 빼서 바꾸기
max_Popul = crime.Population.resample('10YE').max()
crimes_10.Population = max_Popul
display(crimes_10.head())
```

```python
# 미국에 살기 가장 위험했던 연대는?

# 를 확인하기 전에 idxmax() 기능 사용해보기
test_df = pd.DataFrame({'A': [1, 3, 5, 7, 6], 'B': [2, 4, 1, 9, 3]})
print(test_df.idxmax()) # 각 열에 대해 최대값 인덱스 반환
print('-'*50)

# crime 데이터로 확인
crimes_10.idxmax()
# 2000년대가 가장 많음
```
