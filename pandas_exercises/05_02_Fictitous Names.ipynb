```python
# 라이브러리 호출
import pandas as pd

# 데이터 만들기
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)
data3 = pd.DataFrame(raw_data_3)

print(data1.shape, data2.shape, data3.shape)
```

```python
# data1 + data2 세로로 붙여서 'all_data_row'로 만들기
all_data_row = pd.concat([data1, data2])
print(all_data_row.shape)
display(all_data_row)
print('-'*50)

# data1 + data2 가로로 붙여서 'all_data_col'로 만들기
all_data_col = pd.concat([data1, data2], axis=1)
print(all_data_col.shape)
all_data_col
```

```python
# data3의 subject_id와 all_data_row 의 subject_id가 같으니 이를 기준으로 merge 하기
pd.merge(all_data_row, data3, on='subject_id')
```

```python
# data1과 data2를 'subject_id'를 기준으로 inner join 하기
    # how : {'left', 'right', 'outer', 'inner'}, default 'outer'
    #     * left: use only keys from left frame (SQL: left outer join)
    #     * right: use only keys from right frame (SQL: right outer join)
    #     * outer: use union of keys from both frames (SQL: full outer join)
    #     * inner: use intersection of keys from both frames (SQL: inner join).
display(pd.merge(data1, data2, on='subject_id', how='left'))
print('-'*50)
display(pd.merge(data1, data2, on='subject_id', how='right'))
print('-'*50)
display(pd.merge(data1, data2, on='subject_id', how='outer'))
print('-'*50)
display(pd.merge(data1, data2, on='subject_id', how='inner'))
```
