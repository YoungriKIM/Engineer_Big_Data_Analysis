```python
# 라이브러리 호출
import pandas as pd
import numpy as np
from IPython.display import display

csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(csv_url)
print(df.shape)
df.head(3)
```

```python
# school ~ guardian 까지 열만 사용하기(guardian 보호자)
stud_alcoh = df.loc[:, 'school':'guardian']
print(stud_alcoh.shape)
stud_alcoh.head(3)
```

```python
# 첫 글자 대문자로 만드는 함수 만들기
capitalizer = lambda x:x.capitalize()

# Mjob과 Fjob에 적용하기
stud_alcoh.Mjob = stud_alcoh.Mjob.apply(capitalizer)
stud_alcoh.Fjob = stud_alcoh.Fjob.apply(capitalizer)
stud_alcoh.head(3)
```

```python
# 'legal_drinker' 라는 열을 만들기 위한 함수 만들기
# 함수 이름은 'majority'이고 boolean을 반환 17살보다 나이가 많을 때만 합법(True), 어릴때는 불법(False)

# 함수 만들고
def majority(x):
    if x > 17: return True
    else: return False

# 한번 확인하고
print(stud_alcoh.age.apply(majority)[:3])
# 새로운 열 추가하기
stud_alcoh['legal_drinker'] = stud_alcoh.age.apply(majority)
stud_alcoh.head(3)
```

```python
# 숫자 타입의 데이터에는 모두 10 곱하기
stud_alcoh = stud_alcoh.applymap(lambda x: x*10 if isinstance(x, int) else x)
stud_alcoh
```
